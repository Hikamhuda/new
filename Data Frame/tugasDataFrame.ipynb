```python
import pandas as pd
```

```python
df_excel= pd.read_excel('datasampah.xlsx', sheet_name='data')
df_excel
```

```python
total = 0
for i, row in df_excel.iterrows():
  if row['tahun'] == 2016:
    total += row['jumlah_produksi_sampah']
df_pertahun = pd.DataFrame({"tahun": [2016], "total": [total]})
df_pertahun.to_csv("nomor2.csv", index=False)
df_pertahun.to_excel("nomor2.xlsx", index=False)
df_pertahun
```

```python
d_no3 = {}
for i, row in df_excel.iterrows():
  if row['tahun'] not in d_no3:
    d_no3[row['tahun']]=0
  d_no3[row['tahun']] += row['jumlah_produksi_sampah']
df_perkab = pd.DataFrame(d_no3.items(),columns=["tahun", "total"])
df_perkab.to_csv("nomor3.csv", index=False)
df_perkab.to_excel("nomor3.xlsx", index=False)
df_perkab
```

```python
d_no4 = {} 
for i, row in df_excel.iterrows():
    if row['nama_kabupaten_kota'] not in d_no4:
        d_no4[row['nama_kabupaten_kota']] = {}
    if row['tahun'] not in d_no4[row['nama_kabupaten_kota']]:
        d_no4[row['nama_kabupaten_kota']][row['tahun']] = 0
    d_no4[row['nama_kabupaten_kota']][row['tahun']] += row['jumlah_produksi_sampah']
df_perkab_pertahun = pd.DataFrame(d_no4)
df_perkab_pertahun.to_csv("nomor4.csv")
df_perkab_pertahun.to_excel("nomor4.xlsx")
df_perkab_pertahun
```
